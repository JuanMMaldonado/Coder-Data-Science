```python
# Libraries needed for the tutorial

import pandas as pd
import requests
import io
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

#  VINCULO GITHUB Y DEFINICION DE DATA FRAME

```python

# Reading the content and turning it into a pandas dataframe
url = 'https://raw.githubusercontent.com/JuanMMaldonado/Coder-Data-Science/main/Data%20Set%20II%20Ok%20format-%20MRP%20Inventory.csv'
df = pd.read_csv(url)
```

#  GRAFICOS DE MATPLOTLIB



En la siguiente selección se encuentran tres graficos utilizando MATPLOTLIB junto con las preguntas de interes o hipotesis sobre el data set elegido. 

```python
#PREGUNTA1: CUALES SON LAS CANTIDADES TOTALES SEGUN LAS 3 CATEGORIAS EN STOCK?
### -------- 1 -------- ###
#En el siguiente grafico se visualiza las catidades que ya se poseen en el inventario de cada categoria de articulos que se produce. Se puede ver que los articulos de Indumentaria son los más producidos.

# Group the data by division and sum the quantities
grouped = df.groupby('division').sum()
#print(grouped)


# Get the list of categories and the sum of quantities for each category
division = list(grouped.index)
#quantities = list(grouped['qty_inventario'])
#print(quantities)
print(division)
quantities = list(grouped['qty_inventario'])
print(quantities)
# Create the bar chart
plt.bar(division, quantities, width=0.5)

# after plotting the data, format the labels
current_values = plt.gca().get_yticks()

# using format string '{:.0f}' here but you can choose others
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

# Add labels and title to the chart
plt.xlabel("Division")
plt.ylabel("Unidades")
plt.title("RESUMEN")

# Show the chart
plt.show()
```

```python
### -------- 2 -------- ###
# PREGUNTA 2: CUALES SON LAS CANTIDADES TOTALES SEGÚN PAIS DE FABRICACIÓN?
#En el siguiente grafico vemos que cantidades de articulos en total se fabrica en cada pais de producción. Se visualiza que Mexico tiene un nivel mas alto de producción con respectos a los otros paises. 

# Group the data by category and sum the quantities
group1 = df.groupby('pais_fabrica').sum()
#print(group1)
# Get the list of factories and the sum of quantities for each category
fabrica = list(group1.index) 
numero = list(group1['qty_inventario'])
print(numero)
# Create the bar chart
plt.bar(fabrica, numero, width=0.5)

# after plotting the data, format the labels
current_values = plt.gca().get_yticks()

# using format string '{:.0f}' here but you can choose others
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])


# Add labels and title to the chart
plt.xlabel("Fabricas por Pais")
plt.ylabel("Unidades")
plt.title("RESUMEN DE PRODUCCION")


# Rotate X axis legend
plt.xticks(rotation = 45)

# Show the chart
plt.show()
```

```python
#Histograma por categorias
h = sns.histplot(data=df, x="pais_fabrica", hue="division", multiple="stack")

plt.xticks(rotation = 45)
```

```python
### -------- B -------- ###
#PREGUNTA 5: CUALES SON LAS UNIDADES DE PRODUCCION SEGUN DIVISION, PAIS DE FABRICACIÓN Y GÉNERO?
#En el siguiente grafico vemos que cantidades se produjo en los diferentes paises en base al genero del articulo, y dividido segun tipo de division. Se visualiza una producción mayor de articulos para hombre. 

g = sns.displot(data=df, x="pais_fabrica", hue="genero", col="division")

g.set_axis_labels("Fabrica por Pais", "Unidades en Produccion")

g.set_titles("Division: {col_name}")

plt.xticks(rotation = 45)
```

```python
### -------- C -------- ###
#PREGUNTA 6: CUALES SON LAS FECHAS DE ENTREGA DEL FABRICANTE Y SU VOLUMEN DE ENTREGA?
#En el sigiente graficos vemos como se altero el inventario que tenia la empresa en base a las entregas que se iban realizando en fecha y forma.
#Diagrama de dispersión
sns.scatterplot(data=df, x="fecha_entrega", y="qty_inventario")

plt.xticks(rotation = 45)
```

```python
### -------- D -------- ###
#PREGUNTA 7: COMO VARIA DEL PRECIO DE LOS ARTICULOS DE ACUERDO A LAS TEMPORADAS DEL AÑO?
#El siguiente grafico nos muestra como fueron variando los precios por temporada. Vemos las diferencia de precios que tuvieron cada temporada entre si y el valor máximo que llago la ultima temporada (22f)

#Boxplot
ax = sns.boxplot(x="temporada", y="precio", data=df, order=["18F","19S","19F","20S","20F","21S","22F","23S"])
```
